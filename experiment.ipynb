```python
import matplotlib.pyplot as plt
```

```python
file = open('measure_time.txt', 'r')
x = []
y = []
n = 0
while True:
    line = file.readline()
    if not line:
        break
    line = line.split()
   
    x.append(float(float(line[0][:-1])))
    y.append(float(line[1]))
    
    
```

```python
y
```

```python
fig, ax = plt.subplots()
ax.set_title(f"Time of execution and result of function")
ax.set_xlabel('time')
ax.set_ylabel('evalation')
# x = [r[1] for r in result.intermediate_results]
ax.plot(x, y, marker='o')
plt.show()
```
