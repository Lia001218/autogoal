```python
import matplotlib.pyplot as plt
import altair as alt
import pandas as pd
```

```python
file1 = open('measure_time_without.txt', 'r')
x = []
y = []
metalearning = []
file2 = open('measure_time_alternative3.txt', 'r')
while True:
    line = file1.readline()
    if not line:
        break
    line = line.split()
   
    x.append(float(float(line[0][:-1]))/60)
    y.append(float(line[1]))
    metalearning.append(False)
while True:
    line = file2.readline()
    if not line:
        break
    line = line.split()
   
    x.append(float(float(line[0][:-1]))/60)
    y.append(float(line[1]))
    metalearning.append(True)
    
    
```

```python
df = pd.DataFrame({
    'time(min)': x,
    'evaluation': y,
    'metalearning': metalearning
})
```

```python
import altair as alt

alt.Chart(df).mark_circle(size=60).encode(
    x='time(min)',
    y='evaluation',
    color='metalearning',
    tooltip=['time(min)', 'evaluation', 'metalearning']
).interactive()
```
